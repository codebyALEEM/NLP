```python
from gtts import gTTS
from IPython.display import Audio

text_to_speech = gTTS(''' Voice meets code. I integrated gTTS (Google Text-to-Speech) into my Python workflow to transform static text into dynamic audio experiences. With just a few lines of code, I can convert multilingual input into crisp MP3 filesperfect for accessibility features, voice-enabled apps, or chatbot responses. Whether it's English, Hindi, or Spanish, gTTS lets me deliver spoken content with clarity and speed, all powered by Googles robust TTS engine. Its a lightweight yet powerful tool that adds a whole new dimension to user interaction.
''',lang='hi',tld='com')

text_to_speech.save('text_to_speech_gtts.wav')
sound_file = 'text_to_speech_gtts.wav'

Audio(sound_file,autoplay=False)
```

```python
import pyttsx3
from IPython.display import Audio

text = ( """ Innovation isnt just about big ideasits about small, deliberate choices that shape great products.
I code with clarity, design with empathy, and iterate with purpose. Every dropdown, every response, every pixel
is a reflection of how I think: modular, intentional, and user-first. I dont chase trendsI build tools that last,
with clean logic and polished interfaces that speak for themselves. This is where creativity meets precision,
and where every build is a story worth telling.""")

audio = pyttsx3.init()
audio.setProperty('rate',150)
audio.setProperty('volume',0.8)

voice = audio.getProperty('voices')

audio.setProperty('voice',voice[0].id)

audio.say(text)

audio.save_to_file(text,'test_male_Voice.mp3')

audio.runAndWait()
```
