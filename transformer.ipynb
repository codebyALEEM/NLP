```python
from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained("gpt2")

text = "Hello, how are you?"
tokens = tokenizer(text,return_tensors='pt')
print(tokenizer.tokenize(text))
```

```python
from transformers import AutoModelForCausalLM

model = AutoModelForCausalLM.from_pretrained("gpt2")

input_ids = tokenizer.encode("india economy",return_tensors='pt')
output = model.generate(input_ids,max_length=100)
generated_text = tokenizer.decode(output[0],skip_special_tokens=True)

print(generated_text)


```

```python
from transformers import pipeline, set_seed

# Initialize the text generation pipeline with GPT-2
generator = pipeline('text-generation', model='gpt2')

# Set a seed for reproducibility
set_seed(42)

# Generate text based on the IPL cricket prompt
prompt = "Information about IPL cricket matches in India and IPL revenue:"
text_gen = generator(prompt, max_length=30, num_return_sequences=5)

# Print the generated outputs
for i, output in enumerate(text_gen, 1):
    print(f"Generated Text {i}: {output['generated_text']}")
```
